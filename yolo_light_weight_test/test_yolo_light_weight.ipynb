```python
import torch
from torchvision.models import mobilenet_v2

# 1. ‡πÇ‡∏´‡∏•‡∏î "‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏¢‡∏ô‡∏ï‡πå" ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö
mobilenet = mobilenet_v2(weights='IMAGENET1K_V1')
features = mobilenet.features

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á "‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏†‡∏≤‡∏û‡∏à‡∏≥‡∏•‡∏≠‡∏á" ‡∏Ç‡∏ô‡∏≤‡∏î 640x640
dummy_input = torch.randn(1, 3, 640, 640)

print("--- ü©∫ ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡∏ô‡∏¥‡∏à‡∏â‡∏±‡∏¢ MobileNetV2 ---")
print(f"‡∏Ç‡∏ô‡∏≤‡∏î‡∏†‡∏≤‡∏û Input: {dummy_input.shape}\n")

# 3. ‡πÄ‡∏£‡∏¥‡πà‡∏° "‡πÄ‡∏î‡∏¥‡∏ô‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á" ‡∏ó‡∏µ‡∏•‡∏∞‡∏ä‡∏±‡πâ‡∏ô ‡πÅ‡∏•‡πâ‡∏ß‡∏ß‡∏±‡∏î‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
x = dummy_input
for i, module in enumerate(features):
    # ‡∏î‡∏∂‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏•‡∏≤‡∏™‡∏Ç‡∏≠‡∏á‡πÇ‡∏°‡∏î‡∏π‡∏• (Layer)
    module_name = module.__class__.__name__

    # ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì
    x = module(x)

    # ‡∏û‡∏¥‡∏°‡∏û‡πå‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå
    # from: -1 ‡πÄ‡∏™‡∏°‡∏≠ ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÅ‡∏ö‡∏ö Sequential
    print(f"Layer {i:<2}: {module_name:<20} from: -1 \t Output Shape = {x.shape}")
```

```python
import torch
from ultralytics import YOLO

# 1. ‡πÇ‡∏´‡∏•‡∏î‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ X-Ray
model = YOLO('yolo12m.pt')

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á "‡∏†‡∏≤‡∏û‡∏à‡∏≥‡∏•‡∏≠‡∏á" ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å Normalize ‡πÅ‡∏•‡πâ‡∏ß (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÅ‡∏Å‡πâ WARNING)
#    torch.rand ‡∏à‡∏∞‡πÉ‡∏´‡πâ‡∏Ñ‡πà‡∏≤‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 0.0 - 1.0 ‡πÇ‡∏î‡∏¢‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥
dummy_input = torch.rand(1, 3, 640, 640)

# 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á "‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏î‡∏±‡∏Å‡∏ü‡∏±‡∏á" (‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏î‡∏¥‡∏°)
results = []
def get_shape_recursively(output):
    if isinstance(output, torch.Tensor):
        return output.shape
    elif isinstance(output, (list, tuple)):
        return [get_shape_recursively(o) for o in output]
    else:
        return type(output).__name__

def hook_fn(module, input, output):
    results.append({
        'name': module.__class__.__name__,
        'shape': get_shape_recursively(output),
        'from': module.f if hasattr(module, 'f') else -1
    })

# 4. "‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏î‡∏±‡∏Å‡∏ü‡∏±‡∏á" ‡πÅ‡∏•‡∏∞ "‡∏õ‡πâ‡∏≠‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•" ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á (***‡∏™‡πà‡∏ß‡∏ô‡∏ó‡∏µ‡πà‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç***)
print("--- üî¨ ‡∏Å‡∏≥‡∏•‡∏±‡∏á X-Ray ‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏Ç‡∏ô‡∏≤‡∏î Feature Map (‡πÄ‡∏ß‡∏≠‡∏£‡πå‡∏ä‡∏±‡∏ô‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢) ---")

# ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á hook ‡πÅ‡∏Ñ‡πà‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÄ‡∏î‡∏µ‡∏¢‡∏ß ‡πÅ‡∏•‡∏∞‡πÄ‡∏Å‡πá‡∏ö "handle" ‡πÑ‡∏ß‡πâ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ô‡∏≥‡∏≠‡∏≠‡∏Å‡∏ó‡∏µ‡∏´‡∏•‡∏±‡∏á
handles = [m.register_forward_hook(hook_fn) for m in model.model.model]

try:
    # ‡∏õ‡πâ‡∏≠‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏£‡∏∞‡∏ï‡∏∏‡πâ‡∏ô hook ‡πÉ‡∏´‡πâ‡∏ó‡∏≥‡∏á‡∏≤‡∏ô
    _ = model(dummy_input)
finally:
    # ‡∏ô‡∏≥ hook ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏≠‡∏≠‡∏Å‡∏´‡∏•‡∏±‡∏á‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡πÄ‡∏™‡∏£‡πá‡∏à‡πÄ‡∏™‡∏°‡∏≠
    for handle in handles:
        handle.remove()

# 5. "‡∏û‡∏¥‡∏°‡∏û‡πå‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô" (‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏î‡∏¥‡∏°)
for i, result in enumerate(results):
    shape_str = str(result['shape']).replace('torch.Size(', '').replace(')', '')
    print(f"Layer {i:<2}: {result['name']:<10} \t Output Shape: {shape_str}")

    if result['from'] != -1:
        print(f"  ‚îî‚îÄ‚îÄ ‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏≤‡∏à‡∏≤‡∏Å Layer: {result['from']}")
```

```python
import torch
from torchvision.models import mobilenet_v3_small, MobileNet_V3_Small_Weights

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 1: ‡πÇ‡∏´‡∏•‡∏î "‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏¢‡∏ô‡∏ï‡πå" MobileNetV3-Small
# ==============================================================================
print("--- ‚öôÔ∏è ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÇ‡∏´‡∏•‡∏î MobileNetV3-Small ‡∏à‡∏≤‡∏Å torchvision ---")
# ‡πÄ‡∏£‡∏≤‡πÉ‡∏ä‡πâ weights ‡∏ó‡∏µ‡πà‡∏ó‡∏±‡∏ô‡∏™‡∏°‡∏±‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤‡πÑ‡∏î‡πâ‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
mobilenet = mobilenet_v3_small(weights=MobileNet_V3_Small_Weights.DEFAULT)
features = mobilenet.features
print("‚úÖ ‡πÇ‡∏´‡∏•‡∏î‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")


# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 2: ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡∏ô‡∏¥‡∏à‡∏â‡∏±‡∏¢
# ==============================================================================
# ‡∏™‡∏£‡πâ‡∏≤‡∏á "‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏†‡∏≤‡∏û‡∏à‡∏≥‡∏•‡∏≠‡∏á" ‡∏Ç‡∏ô‡∏≤‡∏î 640x640
dummy_input = torch.randn(1, 3, 640, 640)

print("\n\n--- ü©∫ ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡∏ô‡∏¥‡∏à‡∏â‡∏±‡∏¢ MobileNetV3-Small (‡πÄ‡∏ß‡∏≠‡∏£‡πå‡∏ä‡∏±‡∏ô‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå) ---")
print(f"‡∏Ç‡∏ô‡∏≤‡∏î‡∏†‡∏≤‡∏û Input: {dummy_input.shape}\n")

# ‡πÄ‡∏£‡∏¥‡πà‡∏° "‡πÄ‡∏î‡∏¥‡∏ô‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á" ‡∏ó‡∏µ‡∏•‡∏∞‡∏ä‡∏±‡πâ‡∏ô ‡πÅ‡∏•‡πâ‡∏ß‡∏ß‡∏±‡∏î‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
x = dummy_input
for i, module in enumerate(features):
    # ‡∏î‡∏∂‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏•‡∏≤‡∏™‡∏Ç‡∏≠‡∏á‡πÇ‡∏°‡∏î‡∏π‡∏• (Layer)
    module_name = module.__class__.__name__

    # ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì
    x = module(x)

    # ‡∏û‡∏¥‡∏°‡∏û‡πå‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå
    # from: -1 ‡πÄ‡∏™‡∏°‡∏≠ ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÅ‡∏ö‡∏ö Sequential
    print(f"Layer {i:<2}: {module_name:<20} from: -1 \t Output Shape = {x.shape}")
```

```python
import torch
import torch.nn as nn
from ultralytics import YOLO

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 1: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß "yolo-mobilenet-compatible.yaml"
# ==============================================================================
# ‡∏ô‡∏µ‡πà‡∏Ñ‡∏∑‡∏≠‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏≠‡∏≠‡∏Å‡πÅ‡∏ö‡∏ö‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
# - Backbone ‡πÄ‡∏•‡∏µ‡∏¢‡∏ô‡πÅ‡∏ö‡∏ö‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°‡∏Å‡∏≤‡∏£‡∏•‡∏î‡∏Ç‡∏ô‡∏≤‡∏î‡πÅ‡∏•‡∏∞‡πÄ‡∏û‡∏¥‡πà‡∏° channel ‡∏Ç‡∏≠‡∏á Mobilenet-V3-Small
# - Neck ‡πÅ‡∏•‡∏∞ Head ‡∏ñ‡∏π‡∏Å "‡πÄ‡∏î‡∏¥‡∏ô‡∏™‡∏≤‡∏¢‡πÑ‡∏ü" ‡πÉ‡∏´‡πâ‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏ï‡πà‡∏≠‡∏Å‡∏±‡∏ö Backbone ‡πÉ‡∏´‡∏°‡πà‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
yaml_content = """
# Parameters
nc: 80
depth_multiple: 0.50  # ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏õ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ô‡∏µ‡πâ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ï‡∏∑‡πâ‡∏ô‡∏´‡∏£‡∏∑‡∏≠‡∏•‡∏∂‡∏Å‡πÑ‡∏î‡πâ
width_multiple: 1.00   # ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏õ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ô‡∏µ‡πâ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÅ‡∏Ñ‡∏ö‡∏´‡∏£‡∏∑‡∏≠‡∏Å‡∏ß‡πâ‡∏≤‡∏á‡πÑ‡∏î‡πâ
#   n: depth_multiple=0.50, width_multiple=0.25
#   s: depth_multiple=0.50, width_multiple=0.50
#   m: depth_multiple=0.50, width_multiple=1.00
#   l: depth_multiple=1.00, width_multiple=1.00
#   x: depth_multiple=1.00, width_multiple=1.50

# ----------------- 1. BACKBONE (Mobilenet-V3-Small Compatible) -----------------
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏à‡∏≤‡∏Å‡πÇ‡∏°‡∏î‡∏π‡∏•‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á YOLO ‡πÅ‡∏ï‡πà‡πÄ‡∏•‡∏µ‡∏¢‡∏ô‡πÅ‡∏ö‡∏ö‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á MobileNetV2
backbone:
  # [from, repeats, module, args]
  # args: [out_channels, kernel_size, stride]

  # Mimics Layer 0: Conv2dNormActivation -> Output: [16, 320, 320]
  - [-1, 1, Conv, [16, 3, 2]]   # Block 0: Halves the image size from 640 to 320.

  # Mimics Layer 1: InvertedResidual -> Output: [16, 160, 160]
  - [-1, 1, Conv, [16, 3, 2]]   # Block 1: Halves the image size from 320 to 160.

  # Mimics Layer 2: InvertedResidual -> Output: [24, 80, 80]
  - [-1, 1, Conv, [24, 3, 2]]   # Block 2: Halves the image size from 160 to 80. This is the P3 feature map.

  # Mimics Layer 3: InvertedResidual -> Output: [24, 80, 80]
  - [-1, 1, C2f, [24, True]]    # Block 3: Deep feature extraction at the same scale.

  # Mimics Layer 4: InvertedResidual -> Output: [40, 40, 40]
  - [-1, 1, Conv, [40, 3, 2]]   # Block 4: Halves the image size from 80 to 40.

  # Mimics Layer 5 & 6: InvertedResidual -> Output: [40, 40, 40]
  - [-1, 2, C2f, [40, True]]    # Block 5: Two repeats of C2f for deep feature extraction.

  # Mimics Layer 7 & 8: InvertedResidual -> Output: [48, 40, 40]
  - [-1, 1, Conv, [48, 1, 1]]   # Block 6: Pointwise Conv to increase channels from 40 to 48 without changing size.
  - [-1, 1, C2f, [48, True]]    # Block 7: Two repeats of C2f. This is the P4 feature map.

  # Mimics Layer 9: InvertedResidual -> Output: [96, 20, 20]
  - [-1, 1, Conv, [96, 3, 2]]   # Block 8: Halves the image size from 40 to 20.

  # Mimics Layer 10 & 11: InvertedResidual -> Output: [96, 20, 20]
  - [-1, 2, C2f, [96, True]]    # Block 9: Two repeats of C2f for deep feature extraction.

  # Mimics Layer 12: Conv2dNormActivation -> Output: [576, 20, 20]
  - [-1, 1, Conv, [576, 1, 1]]  # Block 10: Pointwise Conv for final channel expansion. This is the P5 feature map.

# ----------------- 2. NECK + HEAD -----------------
# YOLO12n head
head:
  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]          # Block 11
  - [[-1, 7], 1, Concat, [1]] # cat backbone P4         # Block 12
  - [-1, 2, A2C2f, [512, False, -1]]                    # Block 13

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]          # Block 14
  - [[-1, 3], 1, Concat, [1]] # cat backbone P3         # Block 15
  - [-1, 2, A2C2f, [256, False, -1]]                    # Block 16

  - [-1, 1, Conv, [256, 3, 2]]                          # Block 17
  - [[-1, 13], 1, Concat, [1]] # cat head P4            # Block 18
  - [-1, 2, A2C2f, [512, False, -1]]                    # Block 19

  - [-1, 1, Conv, [512, 3, 2]]                          # Block 20
  - [[-1, 10], 1, Concat, [1]] # cat head P5             # Block 21
  - [-1, 2, C3k2, [1024, True]] # 20 (P5/32-large)      # Block 22

  - [[16, 19, 22], 1, Detect, [nc]] # Detect(P3, P4, P5)# Block 23
"""

# ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô string ‡∏•‡∏á‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå
file_name = "yolo-mobilenet-compatible.yaml"
with open(file_name, "w") as f:
    f.write(yaml_content)

print(f"‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß '{file_name}' ‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå‡πÅ‡∏ö‡∏ö‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 2: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏à‡∏≤‡∏Å‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•
# ==============================================================================
print("\n--- üèóÔ∏è ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏à‡∏≤‡∏Å‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢ ---")
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏à‡∏≤‡∏Å‡πÑ‡∏ü‡∏•‡πå YAML ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á 100%
final_model = YOLO(file_name)

# ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡∏≠‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà
final_model.info()

print("\n\nüéâüéâüéâ ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß! üéâüéâüéâ")
print("‡∏ô‡∏µ‡πà‡∏Ñ‡∏∑‡∏≠‡πÇ‡∏°‡πÄ‡∏î‡∏• YOLO ‡∏ó‡∏µ‡πà‡∏°‡∏µ Backbone ‡∏ã‡∏∂‡πà‡∏á‡πÑ‡∏î‡πâ‡∏£‡∏±‡∏ö‡πÅ‡∏£‡∏á‡∏ö‡∏±‡∏ô‡∏î‡∏≤‡∏•‡πÉ‡∏à‡∏à‡∏≤‡∏Å MobileNetV2")


# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 3: ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ X-Ray ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà
# ==============================================================================
print("\n--- üî¨ ‡∏Å‡∏≥‡∏•‡∏±‡∏á X-Ray ‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏Ç‡∏ô‡∏≤‡∏î Feature Map ---")
dummy_input = torch.rand(1, 3, 640, 640)
results = []

def get_shape_recursively(output):
    if isinstance(output, torch.Tensor): return output.shape
    elif isinstance(output, (list, tuple)): return [get_shape_recursively(o) for o in output]
    else: return type(output).__name__

def hook_fn(module, input, output):
    results.append({
        'name': module.__class__.__name__,
        'shape': get_shape_recursively(output),
        'from': module.f if hasattr(module, 'f') else 'N/A'
    })

handles = [m.register_forward_hook(hook_fn) for m in final_model.model.model]
try:
    _ = final_model(dummy_input)
finally:
    for handle in handles:
        handle.remove()

print("\n--- üìù ‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô‡∏ú‡∏•‡∏Å‡∏≤‡∏£ X-Ray ---")
for i, result in enumerate(results):
    shape_str = str(result['shape']).replace('torch.Size(', '').replace(')', '')
    # ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ i ‡∏à‡∏≤‡∏Å enumerate() ‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç‡πÄ‡∏•‡πÄ‡∏¢‡∏≠‡∏£‡πå‡∏ó‡∏µ‡πà‡πÅ‡∏ó‡πâ‡∏à‡∏£‡∏¥‡∏á
    print(f"Layer {i:<2}: {result['name']:<20} \t Output Shape: {shape_str}")
    if result['from'] != -1 and result['from'] != 'N/A':
        print(f"  ‚îî‚îÄ‚îÄ ‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏≤‡∏à‡∏≤‡∏Å Layer: {result['from']}")

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 4: ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Å‡∏±‡∏ö .info(detailed=True)
# ==============================================================================
print("\n\n--- üìä ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Å‡∏±‡∏ö‡∏ú‡∏•‡∏à‡∏≤‡∏Å .info(detailed=True) ---")
final_model.info(detailed=True)

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 5: ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏ü‡∏•‡πå .pt
# ==============================================================================
# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏ü‡∏•‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å
new_model_filename = "yolo_mobilenetv3small_compleate_form.pt"

# ‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á .save() ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏™‡∏ñ‡∏≤‡∏õ‡∏±‡∏ï‡∏¢‡∏Å‡∏£‡∏£‡∏°‡πÅ‡∏•‡∏∞‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ)
final_model.save(new_model_filename)

print(f"\nüíæ ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ñ‡∏π‡∏Å‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏µ‡πà '{new_model_filename}'")
print("‡∏Ñ‡∏∏‡∏ì‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏ô‡∏≥‡πÑ‡∏ü‡∏•‡πå‡∏ô‡∏µ‡πâ‡πÑ‡∏õ‡πÉ‡∏ä‡πâ‡πÄ‡∏ó‡∏£‡∏ô‡∏ï‡πà‡∏≠‡πÑ‡∏î‡πâ‡∏ó‡∏±‡∏ô‡∏ó‡∏µ!")
```

```python
import torch
import torch.nn as nn
from ultralytics import YOLO
from torchvision.models import mobilenet_v2

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 1: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß "yolo-mobilenet-compatible.yaml"
# ==============================================================================
# ‡∏ô‡∏µ‡πà‡∏Ñ‡∏∑‡∏≠‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏≠‡∏≠‡∏Å‡πÅ‡∏ö‡∏ö‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
# - Backbone ‡πÄ‡∏•‡∏µ‡∏¢‡∏ô‡πÅ‡∏ö‡∏ö‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°‡∏Å‡∏≤‡∏£‡∏•‡∏î‡∏Ç‡∏ô‡∏≤‡∏î‡πÅ‡∏•‡∏∞‡πÄ‡∏û‡∏¥‡πà‡∏° channel ‡∏Ç‡∏≠‡∏á MobileNetV2
# - Neck ‡πÅ‡∏•‡∏∞ Head ‡∏ñ‡∏π‡∏Å "‡πÄ‡∏î‡∏¥‡∏ô‡∏™‡∏≤‡∏¢‡πÑ‡∏ü" ‡πÉ‡∏´‡πâ‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏ï‡πà‡∏≠‡∏Å‡∏±‡∏ö Backbone ‡πÉ‡∏´‡∏°‡πà‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
yaml_content = """
# Parameters
nc: 80
depth_multiple: 0.50  # ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏õ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ô‡∏µ‡πâ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ï‡∏∑‡πâ‡∏ô‡∏´‡∏£‡∏∑‡∏≠‡∏•‡∏∂‡∏Å‡πÑ‡∏î‡πâ
width_multiple: 1.00   # ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏õ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ô‡∏µ‡πâ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÅ‡∏Ñ‡∏ö‡∏´‡∏£‡∏∑‡∏≠‡∏Å‡∏ß‡πâ‡∏≤‡∏á‡πÑ‡∏î‡πâ
#   n: depth_multiple=0.50, width_multiple=0.25
#   s: depth_multiple=0.50, width_multiple=0.50
#   m: depth_multiple=0.50, width_multiple=1.00
#   l: depth_multiple=1.00, width_multiple=1.00
#   x: depth_multiple=1.00, width_multiple=1.50

# ----------------- 1. BACKBONE (MobileNetV2 Compatible) -----------------
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏à‡∏≤‡∏Å‡πÇ‡∏°‡∏î‡∏π‡∏•‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á YOLO ‡πÅ‡∏ï‡πà‡πÄ‡∏•‡∏µ‡∏¢‡∏ô‡πÅ‡∏ö‡∏ö‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á MobileNetV2
backbone:
  # [from, repeats, module, args]
  - [-1, 1, Conv, [32, 3, 2]]   # Block 0: Mimics MobileNetV2 Layer 0. Output: [32, 320, 320]
  - [-1, 1, Conv, [16, 1, 1]]   # Block 1: Pointwise Conv to adjust channels. Mimics Layer 1. Output: [16, 320, 320]
  - [-1, 1, Conv, [24, 3, 2]]   # Block 2: Downsample and feature extraction. Mimics Layer 2. Output: [24, 160, 160]
  - [-1, 1, C2f, [24, True]]    # Block 3: Deep feature extraction. Mimics Layer 3.

  # --- P3/8 Feature Extraction ---
  - [-1, 1, Conv, [32, 3, 2]]   # Block 4: Downsample. Mimics Layer 4. Output: [32, 80, 80]
  - [-1, 2, C2f, [32, True]]    # Block 5: Deep feature extraction. This is the P3 feature map for the neck.

  # --- P4/16 Feature Extraction ---
  - [-1, 1, Conv, [64, 3, 2]]   # Block 6: Downsample. Mimics Layer 7. Output: [64, 40, 40]
  - [-1, 3, C2f, [64, True]]    # Block 7: Deep feature extraction. This is the P4 feature map for the neck.
  - [-1, 3, C2f, [96, True]]    # Block 8

  # --- P5/32 Feature Extraction ---
  - [-1, 1, Conv, [160, 3, 2]]  # Block 9: Downsample. Mimics Layer 14. Output: [160, 20, 20]
  - [-1, 2, C2f, [160, True]]   # Block 10: Deep feature extraction. This is the P5 feature map for the neck.
  - [-1, 1, C2f, [320, True]]   # Block 11
  - [-1, 1, C2f, [1280, True]]  # Block 12

# ----------------- 2. NECK + HEAD -----------------
# YOLO12n head
head:
  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]            # Block 13
  - [[-1, 8], 1, Concat, [1]] # cat backbone P4           # Block 14
  - [-1, 2, A2C2f, [512, False, -1]]                      # Block 15

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]            # Block 16
  - [[-1, 5], 1, Concat, [1]] # cat backbone P3           # Block 17
  - [-1, 2, A2C2f, [256, False, -1]]                      # Block 18

  - [-1, 1, Conv, [256, 3, 2]]                            # Block 19
  - [[-1, 15], 1, Concat, [1]] # cat head P4              # Block 20
  - [-1, 2, A2C2f, [512, False, -1]]                      # Block 21

  - [-1, 1, Conv, [512, 3, 2]]                            # Block 22
  - [[-1, 12], 1, Concat, [1]] # cat head P5              # Block 23
  - [-1, 2, C3k2, [1024, True]] # (P5/32-large)           # Block 24

  - [[18, 21, 24], 1, Detect, [nc]] # Detect(P3, P4, P5)  # Block 25
"""

# ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô string ‡∏•‡∏á‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå
file_name = "yolo-mobilenet-compatible.yaml"
with open(file_name, "w") as f:
    f.write(yaml_content)

print(f"‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß '{file_name}' ‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå‡πÅ‡∏ö‡∏ö‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 2: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏à‡∏≤‡∏Å‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•
# ==============================================================================
print("\n--- üèóÔ∏è ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏à‡∏≤‡∏Å‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢ ---")
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏à‡∏≤‡∏Å‡πÑ‡∏ü‡∏•‡πå YAML ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á 100%
final_model = YOLO(file_name)

# ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡∏≠‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà
final_model.info()

print("\n\nüéâüéâüéâ ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß! üéâüéâüéâ")
print("‡∏ô‡∏µ‡πà‡∏Ñ‡∏∑‡∏≠‡πÇ‡∏°‡πÄ‡∏î‡∏• YOLO ‡∏ó‡∏µ‡πà‡∏°‡∏µ Backbone ‡∏ã‡∏∂‡πà‡∏á‡πÑ‡∏î‡πâ‡∏£‡∏±‡∏ö‡πÅ‡∏£‡∏á‡∏ö‡∏±‡∏ô‡∏î‡∏≤‡∏•‡πÉ‡∏à‡∏à‡∏≤‡∏Å MobileNetV2")


# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 3: ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ X-Ray ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà
# ==============================================================================
print("\n--- üî¨ ‡∏Å‡∏≥‡∏•‡∏±‡∏á X-Ray ‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏Ç‡∏ô‡∏≤‡∏î Feature Map ---")
dummy_input = torch.rand(1, 3, 640, 640)
results = []

def get_shape_recursively(output):
    if isinstance(output, torch.Tensor): return output.shape
    elif isinstance(output, (list, tuple)): return [get_shape_recursively(o) for o in output]
    else: return type(output).__name__

def hook_fn(module, input, output):
    results.append({
        'name': module.__class__.__name__,
        'shape': get_shape_recursively(output),
        'from': module.f if hasattr(module, 'f') else 'N/A'
    })

handles = [m.register_forward_hook(hook_fn) for m in final_model.model.model]
try:
    _ = final_model(dummy_input)
finally:
    for handle in handles:
        handle.remove()

print("\n--- üìù ‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô‡∏ú‡∏•‡∏Å‡∏≤‡∏£ X-Ray ---")
for i, result in enumerate(results):
    shape_str = str(result['shape']).replace('torch.Size(', '').replace(')', '')
    # ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ i ‡∏à‡∏≤‡∏Å enumerate() ‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç‡πÄ‡∏•‡πÄ‡∏¢‡∏≠‡∏£‡πå‡∏ó‡∏µ‡πà‡πÅ‡∏ó‡πâ‡∏à‡∏£‡∏¥‡∏á
    print(f"Layer {i:<2}: {result['name']:<20} \t Output Shape: {shape_str}")
    if result['from'] != -1 and result['from'] != 'N/A':
        print(f"  ‚îî‚îÄ‚îÄ ‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏≤‡∏à‡∏≤‡∏Å Layer: {result['from']}")

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 4: ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Å‡∏±‡∏ö .info(detailed=True)
# ==============================================================================
print("\n\n--- üìä ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Å‡∏±‡∏ö‡∏ú‡∏•‡∏à‡∏≤‡∏Å .info(detailed=True) ---")
final_model.info(detailed=True)

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 5: ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏ü‡∏•‡πå .pt
# ==============================================================================
# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏ü‡∏•‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å
new_model_filename = "yolo_mobilenet_compleate_form.pt"

# ‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á .save() ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏™‡∏ñ‡∏≤‡∏õ‡∏±‡∏ï‡∏¢‡∏Å‡∏£‡∏£‡∏°‡πÅ‡∏•‡∏∞‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ)
final_model.save(new_model_filename)

print(f"\nüíæ ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ñ‡∏π‡∏Å‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏µ‡πà '{new_model_filename}'")
print("‡∏Ñ‡∏∏‡∏ì‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏ô‡∏≥‡πÑ‡∏ü‡∏•‡πå‡∏ô‡∏µ‡πâ‡πÑ‡∏õ‡πÉ‡∏ä‡πâ‡πÄ‡∏ó‡∏£‡∏ô‡∏ï‡πà‡∏≠‡πÑ‡∏î‡πâ‡∏ó‡∏±‡∏ô‡∏ó‡∏µ!")
```

```python
# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 1: ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡πÑ‡∏•‡∏ö‡∏£‡∏≤‡∏£‡∏µ‡πÅ‡∏•‡∏∞ Imports
# ==============================================================================
import os
print("‚è≥ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡πÑ‡∏•‡∏ö‡∏£‡∏≤‡∏£‡∏µ ultralytics...")
os.system('pip install -q ultralytics torchvision')
print("‚úÖ ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô!")

import torch
import torch.nn as nn
from ultralytics import YOLO
from torchvision.models import mobilenet_v2

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 2: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡∏î‡∏π‡∏• Backbone ‡πÉ‡∏´‡∏°‡πà (‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î)
# ==============================================================================
# ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á Class ‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏µ‡πà‡∏´‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏° MobileNetV2 ‡πÑ‡∏ß‡πâ
# ‡πÅ‡∏•‡∏∞‡∏™‡∏±‡πà‡∏á‡πÉ‡∏´‡πâ‡∏°‡∏±‡∏ô‡∏™‡πà‡∏á Feature Map ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ 3 ‡∏£‡∏∞‡∏î‡∏±‡∏ö (P3, P4, P5)

class MobileNetV2Backbone(nn.Module):
    def __init__(self, pretrained=True):
        super().__init__()
        # ‡πÇ‡∏´‡∏•‡∏î MobileNetV2 ‡∏ó‡∏µ‡πà train ‡∏°‡∏≤‡πÅ‡∏•‡πâ‡∏ß‡∏ö‡∏ô ImageNet
        mobilenet = mobilenet_v2(weights='IMAGENET1K_V1' if pretrained else None)

        # ‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ Feature Map ‡∏à‡∏≤‡∏Å Layer ‡∏ó‡∏µ‡πà 6, 13, ‡πÅ‡∏•‡∏∞ 17
        self.features = mobilenet.features
        self.feature_indices = (6, 13, 17)

    def forward(self, x):
        # ‡∏™‡∏£‡πâ‡∏≤‡∏á list ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏Å‡πá‡∏ö outputs ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£
        outputs = []
        # ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡∏ú‡πà‡∏≤‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞ layer ‡∏Ç‡∏≠‡∏á MobileNetV2
        for i, module in enumerate(self.features):
            x = module(x)
            # ‡∏ñ‡πâ‡∏≤‡πÄ‡∏à‡∏≠ layer ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ ‡πÉ‡∏´‡πâ‡πÄ‡∏Å‡πá‡∏ö output ‡πÑ‡∏ß‡πâ
            if i in self.feature_indices:
                outputs.append(x)
        return outputs

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 3: ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ "‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î" ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô Backbone
# ==============================================================================
print("\n--- ü©∫ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏Å‡∏≤‡∏£‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô Backbone ---")

# 1. ‡πÇ‡∏´‡∏•‡∏î‡πÇ‡∏°‡πÄ‡∏î‡∏• YOLOv12m ‡πÄ‡∏î‡∏¥‡∏°
yolo_model = YOLO('yolo12m.pt')
original_layers = yolo_model.model.model

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á Backbone ‡πÉ‡∏´‡∏°‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤
new_backbone = MobileNetV2Backbone()

# 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà‡πÇ‡∏î‡∏¢‡πÄ‡∏≠‡∏≤ Backbone ‡πÉ‡∏´‡∏°‡πà‡πÑ‡∏õ‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà‡∏™‡πà‡∏ß‡∏ô Backbone ‡πÄ‡∏î‡∏¥‡∏°
#    Backbone ‡πÄ‡∏î‡∏¥‡∏°‡∏Ç‡∏≠‡∏á YOLOv12m ‡∏Ñ‡∏∑‡∏≠ 10 layers ‡πÅ‡∏£‡∏Å (index 0-9)
#    ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà 10 layers ‡∏ô‡∏µ‡πâ‡∏î‡πâ‡∏ß‡∏¢ new_backbone ‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏ä‡∏¥‡πâ‡∏ô‡πÄ‡∏î‡∏µ‡∏¢‡∏ß
new_layers = nn.Sequential(
    new_backbone,  # Backbone ‡πÉ‡∏´‡∏°‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤
    *list(original_layers.children())[10:] # ‡πÄ‡∏≠‡∏≤ Neck + Head ‡πÄ‡∏î‡∏¥‡∏°‡∏°‡∏≤‡∏ï‡πà‡∏≠‡∏ó‡πâ‡∏≤‡∏¢
)

# 4. ‡∏ô‡∏≥‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÉ‡∏´‡∏°‡πà‡πÉ‡∏™‡πà‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ‡πÉ‡∏ô‡πÇ‡∏°‡πÄ‡∏î‡∏• YOLO
yolo_model.model.model = new_layers

# 5. "‡πÄ‡∏î‡∏¥‡∏ô‡∏™‡∏≤‡∏¢‡πÑ‡∏ü‡πÉ‡∏´‡∏°‡πà" ‡πÉ‡∏´‡πâ‡∏Å‡∏±‡∏ö Head (‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç `IndexError` ‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà)
#    ‡∏´‡∏≤ Head (‡πÇ‡∏°‡∏î‡∏π‡∏• Detect) ‡∏ã‡∏∂‡πà‡∏á‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡∏à‡∏∞‡∏≠‡∏¢‡∏π‡πà‡∏ó‡∏µ‡πà Layer ‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
detect_head = yolo_model.model.model[-1]

#    Backbone ‡πÉ‡∏´‡∏°‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤‡∏™‡πà‡∏á Output ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ 3 ‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å Layer ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß (Layer 0)
#    ‡∏î‡∏±‡∏á‡∏ô‡∏±‡πâ‡∏ô‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏ö‡∏≠‡∏Å‡πÉ‡∏´‡πâ Head ‡πÑ‡∏õ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å Layer 0 ‡∏ó‡∏±‡πâ‡∏á 3 ‡∏Ñ‡πà‡∏≤
#    (‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏≤‡∏Å PyTorch ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏î‡∏∂‡∏á‡∏´‡∏•‡∏≤‡∏¢ output ‡∏à‡∏≤‡∏Å layer ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡πÑ‡∏î‡πâ‡πÇ‡∏î‡∏¢‡∏ï‡∏£‡∏á
#     ultralytics ‡∏à‡∏∞‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡πÉ‡∏´‡πâ‡πÄ‡∏≠‡∏á‡∏ñ‡πâ‡∏≤‡πÄ‡∏£‡∏≤‡∏ö‡∏≠‡∏Å‡∏ß‡πà‡∏≤‡πÉ‡∏´‡πâ‡∏î‡∏∂‡∏á‡∏à‡∏≤‡∏Å layer 0)
#    ‡∏Å‡∏≤‡∏£‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤ f=[0, 0, 0] ‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡∏¥‡∏ò‡∏µ‡∏ö‡∏≠‡∏Å‡πÉ‡∏´‡πâ‡∏£‡∏∞‡∏ö‡∏ö‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏ô‡∏µ‡πâ
#    ‡πÉ‡∏ô‡∏ó‡∏≤‡∏á‡∏õ‡∏è‡∏¥‡∏ö‡∏±‡∏ï‡∏¥ ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏Å‡∏±‡∏ö layer ‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤ Neck ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡∏´‡∏ô‡πâ‡∏≤‡∏ó‡∏µ‡πà‡πÅ‡∏¢‡∏Å feature
#    ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡∏á‡πà‡∏≤‡∏¢ ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏Å‡∏±‡∏ö layer ‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á Neck ‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤ Head
#    ‡πÉ‡∏ô‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á YOLOv8/v12 ‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏Ñ‡∏∑‡∏≠ layer ‡∏ó‡∏µ‡πà 15, 18, 21
detect_head.f = [15, 18, 21]

print("‚úÖ ‡∏Å‡∏≤‡∏£‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 4: ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
# ==============================================================================
print("\n\n--- üéâ ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏≤‡∏£‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î ---")
yolo_model.info(detailed=True)
print("\n\n‡∏Ñ‡∏∏‡∏ì‡πÑ‡∏î‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏• YOLO ‡∏ó‡∏µ‡πà‡∏°‡∏µ MobileNetV2 ‡πÄ‡∏õ‡πá‡∏ô Backbone ‡πÑ‡∏î‡πâ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß!")

# (‡πÉ‡∏™‡πà‡∏ï‡πà‡∏≠‡∏à‡∏≤‡∏Å‡πÇ‡∏Ñ‡πâ‡∏î‡πÄ‡∏î‡∏¥‡∏°‡∏ó‡∏µ‡πà‡∏Ñ‡∏∏‡∏ì‡∏£‡∏±‡∏ô‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß)

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 5: ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏ü‡∏•‡πå .pt
# ==============================================================================
# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà
new_model_filename = "yolo12m_mobilenetv2.pt"

# ‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á .save() ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•
yolo_model.save(new_model_filename)

print(f"\n\n‚úÖ ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ñ‡∏π‡∏Å‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏µ‡πà '{new_model_filename}'")
```

‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ô‡∏µ‡πâ index error ‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏û‡∏£‡∏≤‡∏∞ ultralytics ‡πÑ‡∏°‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à torchvision

```python
import torch
import torch.nn as nn
from ultralytics import YOLO
from torchvision.models import mobilenet_v2

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 1: ‡∏™‡∏£‡πâ‡∏≤‡∏á Backbone ‡∏ó‡∏µ‡πà‡∏°‡∏µ "‡∏´‡∏°‡πâ‡∏≠‡πÅ‡∏õ‡∏•‡∏á" (Adapter) ‡πÉ‡∏ô‡∏ï‡∏±‡∏ß
# ==============================================================================
class AdaptedMobileNetV2Backbone(nn.Module):
    def __init__(self, pretrained=True):
        super().__init__()
        mobilenet = mobilenet_v2(weights='IMAGENET1K_V1' if pretrained else None)
        self.features = mobilenet.features
        self.feature_indices = (6, 13, 17)
        yolo_channels_needed = [512, 512, 1024]
        mobilenet_channels_out = [32, 96, 320]
        self.adapters = nn.ModuleList([
            nn.Conv2d(in_ch, out_ch, kernel_size=1)
            for in_ch, out_ch in zip(mobilenet_channels_out, yolo_channels_needed)
        ])

    def forward(self, x):
        outputs = []
        mobilenet_outputs = []
        for i, module in enumerate(self.features):
            x = module(x)
            if i in self.feature_indices:
                mobilenet_outputs.append(x)
        for i, adapter in enumerate(self.adapters):
            outputs.append(adapter(mobilenet_outputs[i]))
        return outputs

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 2: ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ "‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î" ‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° "‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ä‡∏¥‡πâ‡∏ô‡∏™‡πà‡∏ß‡∏ô"
# ==============================================================================
print("--- ü©∫ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏• YOLO ‡∏ó‡∏µ‡πà‡∏°‡∏µ MobileNetV2 ‡∏à‡∏£‡∏¥‡∏á‡πÜ ---")

# 1. ‡πÇ‡∏´‡∏•‡∏î‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÄ‡∏î‡∏¥‡∏°‡πÄ‡∏û‡∏∑‡πà‡∏≠ "‡∏Ñ‡∏±‡∏î‡∏•‡∏≠‡∏Å" ‡∏™‡πà‡∏ß‡∏ô Neck+Head ‡πÄ‡∏Å‡πá‡∏ö‡πÑ‡∏ß‡πâ
yolo_original = YOLO('yolo12m.pt')
original_neck_head = list(yolo_original.model.model.children())[10:]

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô ‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î
final_model = YOLO('yolo12m.pt')
new_adapted_backbone = AdaptedMobileNetV2Backbone()
final_model.model.backbone = new_adapted_backbone
print("‚úÖ ‡∏Å‡∏≤‡∏£‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")

# 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á "‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ä‡∏¥‡πâ‡∏ô‡∏™‡πà‡∏ß‡∏ô‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á" ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•
modules_to_display = [final_model.model.backbone] + original_neck_head


# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 3: X-Ray ‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÉ‡∏´‡∏°‡πà (***‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î***)
# ==============================================================================
print("\n--- üî¨ ‡∏Å‡∏≥‡∏•‡∏±‡∏á X-Ray ‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏Ç‡∏≠‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà ---")
dummy_input = torch.rand(1, 3, 640, 640)

# 1. ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö "‡∏ú‡∏•‡∏ú‡∏•‡∏¥‡∏ï" ‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å Backbone ‡πÉ‡∏´‡∏°‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤‡πÇ‡∏î‡∏¢‡∏ï‡∏£‡∏á
backbone_output = final_model.model.backbone(dummy_input)
backbone_output_shapes = [o.shape for o in backbone_output]

# 2. ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö "‡∏ú‡∏•‡∏ú‡∏•‡∏¥‡∏ï‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢" ‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
final_output = final_model(dummy_input)

print("\n--- üìù ‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô‡∏ú‡∏•‡∏Å‡∏≤‡∏£ X-Ray (‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏ö‡∏•‡πá‡∏≠‡∏Å) ---")
# ‡∏û‡∏¥‡∏°‡∏û‡πå "‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ä‡∏¥‡πâ‡∏ô‡∏™‡πà‡∏ß‡∏ô" ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡πÉ‡∏ä‡πâ‡∏õ‡∏£‡∏∞‡∏Å‡∏≠‡∏ö‡πÇ‡∏°‡πÄ‡∏î‡∏•
for i, module in enumerate(modules_to_display):
    module_name = module.__class__.__name__
    if i == 0: # ‡∏ñ‡πâ‡∏≤‡πÄ‡∏õ‡πá‡∏ô Backbone ‡πÉ‡∏´‡∏°‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤
        shape_str = str(backbone_output_shapes).replace('torch.Size(', '').replace(')', '')
        print(f"Block {i:<2}: {module_name:<25} \t Output Shape: {shape_str}")
    else: # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö Neck ‡πÅ‡∏•‡∏∞ Head ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÅ‡∏™‡∏î‡∏á‡πÅ‡∏Ñ‡πà‡∏ä‡∏∑‡πà‡∏≠‡∏ö‡∏•‡πá‡∏≠‡∏Å‡πÅ‡∏•‡∏∞‡∏™‡∏≤‡∏¢‡πÑ‡∏ü
        from_str = f"\n  ‚îî‚îÄ‚îÄ ‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏≤‡∏à‡∏≤‡∏Å Layer: {module.f}" if hasattr(module, 'f') else ""
        print(f"Block {i:<2}: {module_name:<25} \t {from_str}")


# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 4: ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• .info() ‡∏ã‡∏∂‡πà‡∏á‡∏à‡∏∞‡πÅ‡∏™‡∏î‡∏á "‡∏ä‡∏¥‡πâ‡∏ô‡∏™‡πà‡∏ß‡∏ô‡∏¢‡πà‡∏≠‡∏¢" ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
# ==============================================================================
print("\n\n--- üéâ ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏≤‡∏£‡∏ú‡πà‡∏≤‡∏ï‡∏±‡∏î (‡∏°‡∏∏‡∏°‡∏°‡∏≠‡∏á‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏ä‡∏¥‡πâ‡∏ô‡∏™‡πà‡∏ß‡∏ô‡∏¢‡πà‡∏≠‡∏¢) ---")
final_model.info(detailed=True)
print("\n\nüéâüéâüéâ ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à! ‡∏Ñ‡∏∏‡∏ì‡πÑ‡∏î‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö YOLO ‡∏î‡πâ‡∏ß‡∏¢ MobileNetV2 ‡∏à‡∏£‡∏¥‡∏á‡πÜ ‡πÑ‡∏î‡πâ‡πÅ‡∏•‡πâ‡∏ß! üéâüéâüéâ")

# ==============================================================================
#  ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 5: ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏ü‡∏•‡πå .pt (***‡∏™‡πà‡∏ß‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤***)
# ==============================================================================
new_model_filename = "yolo12m_adapted_mobilenetv2.pt"
final_model.save(new_model_filename)
print(f"\n\nüíæ ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÑ‡∏Æ‡∏ö‡∏£‡∏¥‡∏î‡∏ñ‡∏π‡∏Å‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏µ‡πà '{new_model_filename}'")
print("‡∏Ñ‡∏∏‡∏ì‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏ô‡∏≥‡πÑ‡∏ü‡∏•‡πå‡∏ô‡∏µ‡πâ‡πÑ‡∏õ‡πÉ‡∏ä‡πâ‡πÄ‡∏ó‡∏£‡∏ô‡∏ï‡πà‡∏≠‡πÑ‡∏î‡πâ‡∏ó‡∏±‡∏ô‡∏ó‡∏µ!")
```
