```python
import torch
from ultralytics import YOLO

# 1. โหลดโมเดลที่เราต้องการ X-Ray
model = YOLO('yolo12m.pt')

# 2. สร้าง "ภาพจำลอง" ที่ถูก Normalize แล้ว (เพื่อแก้ WARNING)
#    torch.rand จะให้ค่าระหว่าง 0.0 - 1.0 โดยอัตโนมัติ
dummy_input = torch.rand(1, 3, 640, 640)

# 3. สร้าง "เครื่องดักฟัง" (เหมือนเดิม)
results = []
def get_shape_recursively(output):
    if isinstance(output, torch.Tensor):
        return output.shape
    elif isinstance(output, (list, tuple)):
        return [get_shape_recursively(o) for o in output]
    else:
        return type(output).__name__

def hook_fn(module, input, output):
    results.append({
        'name': module.__class__.__name__,
        'shape': get_shape_recursively(output),
        'from': module.f if hasattr(module, 'f') else -1
    })

# 4. "ติดตั้งเครื่องดักฟัง" และ "ป้อนข้อมูล" อย่างถูกต้อง (***ส่วนที่แก้ไข***)
print("--- 🔬 กำลัง X-Ray โครงสร้างและขนาด Feature Map (เวอร์ชันสุดท้าย) ---")

# ติดตั้ง hook แค่ครั้งเดียว และเก็บ "handle" ไว้เพื่อนำออกทีหลัง
handles = [m.register_forward_hook(hook_fn) for m in model.model.model]

try:
    # ป้อนข้อมูลเพื่อกระตุ้น hook ให้ทำงาน
    _ = model(dummy_input)
finally:
    # นำ hook ทั้งหมดออกหลังใช้งานเสร็จเสมอ
    for handle in handles:
        handle.remove()

# 5. "พิมพ์รายงาน" (เหมือนเดิม)
for i, result in enumerate(results):
    shape_str = str(result['shape']).replace('torch.Size(', '').replace(')', '')
    print(f"Layer {i:<2}: {result['name']:<10} \t Output Shape: {shape_str}")

    if result['from'] != -1:
        print(f"  └── รับข้อมูลมาจาก Layer: {result['from']}")
```

## สร้างโดย yaml แรงบันดาลใจจาก Mobilenet-V3-Small สำเร็จ รอเทรน

```python
import torch
import torch.nn as nn
from ultralytics import YOLO

# ==============================================================================
#  ขั้นตอนที่ 1: สร้างพิมพ์เขียว "yolo-mobilenet-compatible.yaml"
# ==============================================================================
# นี่คือพิมพ์เขียวที่เราออกแบบขึ้นมาใหม่ทั้งหมด
# - Backbone เลียนแบบพฤติกรรมการลดขนาดและเพิ่ม channel ของ Mobilenet-V3-Small
# - Neck และ Head ถูก "เดินสายไฟ" ให้เชื่อมต่อกับ Backbone ใหม่อย่างถูกต้อง
yaml_content = """
# Parameters
nc: 80
depth_multiple: 0.50  # สามารถปรับค่านี้เพื่อทำให้โมเดลตื้นหรือลึกได้
width_multiple: 1.00   # สามารถปรับค่านี้เพื่อทำให้โมเดลแคบหรือกว้างได้
#   n: depth_multiple=0.50, width_multiple=0.25
#   s: depth_multiple=0.50, width_multiple=0.50
#   m: depth_multiple=0.50, width_multiple=1.00
#   l: depth_multiple=1.00, width_multiple=1.00
#   x: depth_multiple=1.00, width_multiple=1.50

# ----------------- 1. BACKBONE (Mobilenet-V3-Small Compatible) -----------------
# สร้างจากโมดูลมาตรฐานของ YOLO แต่เลียนแบบโครงสร้าง MobileNetV2
backbone:
  # [from, repeats, module, args]
  # args: [out_channels, kernel_size, stride]

  # Mimics Layer 0: Conv2dNormActivation -> Output: [16, 320, 320]
  - [-1, 1, Conv, [16, 3, 2]]   # Block 0: Halves the image size from 640 to 320.

  # Mimics Layer 1: InvertedResidual -> Output: [16, 160, 160]
  - [-1, 1, Conv, [16, 3, 2]]   # Block 1: Halves the image size from 320 to 160.

  # Mimics Layer 2: InvertedResidual -> Output: [24, 80, 80]
  - [-1, 1, Conv, [24, 3, 2]]   # Block 2: Halves the image size from 160 to 80. This is the P3 feature map.

  # Mimics Layer 3: InvertedResidual -> Output: [24, 80, 80]
  - [-1, 1, C2f, [24, True]]    # Block 3: Deep feature extraction at the same scale.

  # Mimics Layer 4: InvertedResidual -> Output: [40, 40, 40]
  - [-1, 1, Conv, [40, 3, 2]]   # Block 4: Halves the image size from 80 to 40.

  # Mimics Layer 5 & 6: InvertedResidual -> Output: [40, 40, 40]
  - [-1, 2, C2f, [40, True]]    # Block 5: Two repeats of C2f for deep feature extraction.

  # Mimics Layer 7 & 8: InvertedResidual -> Output: [48, 40, 40]
  - [-1, 1, Conv, [48, 1, 1]]   # Block 6: Pointwise Conv to increase channels from 40 to 48 without changing size.
  - [-1, 1, C2f, [48, True]]    # Block 7: Two repeats of C2f. This is the P4 feature map.

  # Mimics Layer 9: InvertedResidual -> Output: [96, 20, 20]
  - [-1, 1, Conv, [96, 3, 2]]   # Block 8: Halves the image size from 40 to 20.

  # Mimics Layer 10 & 11: InvertedResidual -> Output: [96, 20, 20]
  - [-1, 2, C2f, [96, True]]    # Block 9: Two repeats of C2f for deep feature extraction.

  # Mimics Layer 12: Conv2dNormActivation -> Output: [576, 20, 20]
  - [-1, 1, Conv, [576, 1, 1]]  # Block 10: Pointwise Conv for final channel expansion. This is the P5 feature map.

# ----------------- 2. NECK + HEAD -----------------
# YOLO12n head
head:
  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]          # Block 11
  - [[-1, 7], 1, Concat, [1]] # cat backbone P4         # Block 12
  - [-1, 2, A2C2f, [512, False, -1]]                    # Block 13

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]          # Block 14
  - [[-1, 3], 1, Concat, [1]] # cat backbone P3         # Block 15
  - [-1, 2, A2C2f, [256, False, -1]]                    # Block 16

  - [-1, 1, Conv, [256, 3, 2]]                          # Block 17
  - [[-1, 13], 1, Concat, [1]] # cat head P4            # Block 18
  - [-1, 2, A2C2f, [512, False, -1]]                    # Block 19

  - [-1, 1, Conv, [512, 3, 2]]                          # Block 20
  - [[-1, 10], 1, Concat, [1]] # cat head P5             # Block 21
  - [-1, 2, C3k2, [1024, True]] # 20 (P5/32-large)      # Block 22

  - [[16, 19, 22], 1, Detect, [nc]] # Detect(P3, P4, P5)# Block 23
"""

# เขียน string ลงในไฟล์
file_name = "yolo-mobilenet-compatible.yaml"
with open(file_name, "w") as f:
    f.write(yaml_content)

print(f"✅ สร้างพิมพ์เขียว '{file_name}' ที่สมบูรณ์แบบสำเร็จ!")

# ==============================================================================
#  ขั้นตอนที่ 2: สร้างโมเดลจากพิมพ์เขียวและแสดงผล
# ==============================================================================
print("\n--- 🏗️ กำลังสร้างโมเดลจากพิมพ์เขียวสุดท้าย ---")
# สร้างโมเดลจากไฟล์ YAML ที่ถูกต้อง 100%
final_model = YOLO(file_name)

# แสดงข้อมูลสรุปของโมเดลใหม่
final_model.info()

print("\n\n🎉🎉🎉 โมเดลสร้างสำเร็จแล้ว! 🎉🎉🎉")
print("นี่คือโมเดล YOLO ที่มี Backbone ซึ่งได้รับแรงบันดาลใจจาก MobileNetV2")


# ==============================================================================
#  ขั้นตอนที่ 3: ทำการ X-Ray โมเดลใหม่
# ==============================================================================
print("\n--- 🔬 กำลัง X-Ray โครงสร้างและขนาด Feature Map ---")
dummy_input = torch.rand(1, 3, 640, 640)
results = []

def get_shape_recursively(output):
    if isinstance(output, torch.Tensor): return output.shape
    elif isinstance(output, (list, tuple)): return [get_shape_recursively(o) for o in output]
    else: return type(output).__name__

def hook_fn(module, input, output):
    results.append({
        'name': module.__class__.__name__,
        'shape': get_shape_recursively(output),
        'from': module.f if hasattr(module, 'f') else 'N/A'
    })

handles = [m.register_forward_hook(hook_fn) for m in final_model.model.model]
try:
    _ = final_model(dummy_input)
finally:
    for handle in handles:
        handle.remove()

print("\n--- 📝 รายงานผลการ X-Ray ---")
for i, result in enumerate(results):
    shape_str = str(result['shape']).replace('torch.Size(', '').replace(')', '')
    # เราจะใช้ i จาก enumerate() เป็นหมายเลขเลเยอร์ที่แท้จริง
    print(f"Layer {i:<2}: {result['name']:<20} \t Output Shape: {shape_str}")
    if result['from'] != -1 and result['from'] != 'N/A':
        print(f"  └── รับข้อมูลมาจาก Layer: {result['from']}")

# ==============================================================================
#  ขั้นตอนที่ 4: เปรียบเทียบกับ .info(detailed=True)
# ==============================================================================
print("\n\n--- 📊 เปรียบเทียบกับผลจาก .info(detailed=True) ---")
final_model.info(detailed=True)

# ==============================================================================
#  ขั้นตอนที่ 5: บันทึกโมเดลใหม่เป็นไฟล์ .pt
# ==============================================================================
# กำหนดชื่อไฟล์สำหรับโมเดลที่จะบันทึก
new_model_filename = "yolo_mobilenetv3small_compleate_form.pt"

# ใช้คำสั่ง .save() เพื่อบันทึกสถาปัตยกรรมและน้ำหนัก (ถ้ามี)
final_model.save(new_model_filename)

print(f"\n💾 โมเดลถูกบันทึกเรียบร้อยแล้วที่ '{new_model_filename}'")
print("คุณสามารถนำไฟล์นี้ไปใช้เทรนต่อได้ทันที!")
```

## สร้างโดย yaml แรงบันดาลใจจาก Mobilenet-V2 ใหญ่ไป

```python
import torch
import torch.nn as nn
from ultralytics import YOLO
from torchvision.models import mobilenet_v2

# ==============================================================================
#  ขั้นตอนที่ 1: สร้างพิมพ์เขียว "yolo-mobilenet-compatible.yaml"
# ==============================================================================
# นี่คือพิมพ์เขียวที่เราออกแบบขึ้นมาใหม่ทั้งหมด
# - Backbone เลียนแบบพฤติกรรมการลดขนาดและเพิ่ม channel ของ MobileNetV2
# - Neck และ Head ถูก "เดินสายไฟ" ให้เชื่อมต่อกับ Backbone ใหม่อย่างถูกต้อง
yaml_content = """
# Parameters
nc: 80
depth_multiple: 0.50  # สามารถปรับค่านี้เพื่อทำให้โมเดลตื้นหรือลึกได้
width_multiple: 1.00   # สามารถปรับค่านี้เพื่อทำให้โมเดลแคบหรือกว้างได้
#   n: depth_multiple=0.50, width_multiple=0.25
#   s: depth_multiple=0.50, width_multiple=0.50
#   m: depth_multiple=0.50, width_multiple=1.00
#   l: depth_multiple=1.00, width_multiple=1.00
#   x: depth_multiple=1.00, width_multiple=1.50

# ----------------- 1. BACKBONE (MobileNetV2 Compatible) -----------------
# สร้างจากโมดูลมาตรฐานของ YOLO แต่เลียนแบบโครงสร้าง MobileNetV2
backbone:
  # [from, repeats, module, args]
  - [-1, 1, Conv, [32, 3, 2]]   # Block 0: Mimics MobileNetV2 Layer 0. Output: [32, 320, 320]
  - [-1, 1, Conv, [16, 1, 1]]   # Block 1: Pointwise Conv to adjust channels. Mimics Layer 1. Output: [16, 320, 320]
  - [-1, 1, Conv, [24, 3, 2]]   # Block 2: Downsample and feature extraction. Mimics Layer 2. Output: [24, 160, 160]
  - [-1, 1, C2f, [24, True]]    # Block 3: Deep feature extraction. Mimics Layer 3.

  # --- P3/8 Feature Extraction ---
  - [-1, 1, Conv, [32, 3, 2]]   # Block 4: Downsample. Mimics Layer 4. Output: [32, 80, 80]
  - [-1, 2, C2f, [32, True]]    # Block 5: Deep feature extraction. This is the P3 feature map for the neck.

  # --- P4/16 Feature Extraction ---
  - [-1, 1, Conv, [64, 3, 2]]   # Block 6: Downsample. Mimics Layer 7. Output: [64, 40, 40]
  - [-1, 3, C2f, [64, True]]    # Block 7: Deep feature extraction. This is the P4 feature map for the neck.
  - [-1, 3, C2f, [96, True]]    # Block 8

  # --- P5/32 Feature Extraction ---
  - [-1, 1, Conv, [160, 3, 2]]  # Block 9: Downsample. Mimics Layer 14. Output: [160, 20, 20]
  - [-1, 2, C2f, [160, True]]   # Block 10: Deep feature extraction. This is the P5 feature map for the neck.
  - [-1, 1, C2f, [320, True]]   # Block 11
  - [-1, 1, C2f, [1280, True]]  # Block 12

# ----------------- 2. NECK + HEAD -----------------
# YOLO12n head
head:
  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]            # Block 13
  - [[-1, 8], 1, Concat, [1]] # cat backbone P4           # Block 14
  - [-1, 2, A2C2f, [512, False, -1]]                      # Block 15

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]            # Block 16
  - [[-1, 5], 1, Concat, [1]] # cat backbone P3           # Block 17
  - [-1, 2, A2C2f, [256, False, -1]]                      # Block 18

  - [-1, 1, Conv, [256, 3, 2]]                            # Block 19
  - [[-1, 15], 1, Concat, [1]] # cat head P4              # Block 20
  - [-1, 2, A2C2f, [512, False, -1]]                      # Block 21

  - [-1, 1, Conv, [512, 3, 2]]                            # Block 22
  - [[-1, 12], 1, Concat, [1]] # cat head P5              # Block 23
  - [-1, 2, C3k2, [1024, True]] # (P5/32-large)           # Block 24

  - [[18, 21, 24], 1, Detect, [nc]] # Detect(P3, P4, P5)  # Block 25
"""

# เขียน string ลงในไฟล์
file_name = "yolo-mobilenet-compatible.yaml"
with open(file_name, "w") as f:
    f.write(yaml_content)

print(f"✅ สร้างพิมพ์เขียว '{file_name}' ที่สมบูรณ์แบบสำเร็จ!")

# ==============================================================================
#  ขั้นตอนที่ 2: สร้างโมเดลจากพิมพ์เขียวและแสดงผล
# ==============================================================================
print("\n--- 🏗️ กำลังสร้างโมเดลจากพิมพ์เขียวสุดท้าย ---")
# สร้างโมเดลจากไฟล์ YAML ที่ถูกต้อง 100%
final_model = YOLO(file_name)

# แสดงข้อมูลสรุปของโมเดลใหม่
final_model.info()

print("\n\n🎉🎉🎉 โมเดลสร้างสำเร็จแล้ว! 🎉🎉🎉")
print("นี่คือโมเดล YOLO ที่มี Backbone ซึ่งได้รับแรงบันดาลใจจาก MobileNetV2")


# ==============================================================================
#  ขั้นตอนที่ 3: ทำการ X-Ray โมเดลใหม่
# ==============================================================================
print("\n--- 🔬 กำลัง X-Ray โครงสร้างและขนาด Feature Map ---")
dummy_input = torch.rand(1, 3, 640, 640)
results = []

def get_shape_recursively(output):
    if isinstance(output, torch.Tensor): return output.shape
    elif isinstance(output, (list, tuple)): return [get_shape_recursively(o) for o in output]
    else: return type(output).__name__

def hook_fn(module, input, output):
    results.append({
        'name': module.__class__.__name__,
        'shape': get_shape_recursively(output),
        'from': module.f if hasattr(module, 'f') else 'N/A'
    })

handles = [m.register_forward_hook(hook_fn) for m in final_model.model.model]
try:
    _ = final_model(dummy_input)
finally:
    for handle in handles:
        handle.remove()

print("\n--- 📝 รายงานผลการ X-Ray ---")
for i, result in enumerate(results):
    shape_str = str(result['shape']).replace('torch.Size(', '').replace(')', '')
    # เราจะใช้ i จาก enumerate() เป็นหมายเลขเลเยอร์ที่แท้จริง
    print(f"Layer {i:<2}: {result['name']:<20} \t Output Shape: {shape_str}")
    if result['from'] != -1 and result['from'] != 'N/A':
        print(f"  └── รับข้อมูลมาจาก Layer: {result['from']}")

# ==============================================================================
#  ขั้นตอนที่ 4: เปรียบเทียบกับ .info(detailed=True)
# ==============================================================================
print("\n\n--- 📊 เปรียบเทียบกับผลจาก .info(detailed=True) ---")
final_model.info(detailed=True)

# ==============================================================================
#  ขั้นตอนที่ 5: บันทึกโมเดลใหม่เป็นไฟล์ .pt
# ==============================================================================
# กำหนดชื่อไฟล์สำหรับโมเดลที่จะบันทึก
new_model_filename = "yolo_mobilenet_compleate_form.pt"

# ใช้คำสั่ง .save() เพื่อบันทึกสถาปัตยกรรมและน้ำหนัก (ถ้ามี)
final_model.save(new_model_filename)

print(f"\n💾 โมเดลถูกบันทึกเรียบร้อยแล้วที่ '{new_model_filename}'")
print("คุณสามารถนำไฟล์นี้ไปใช้เทรนต่อได้ทันที!")
```

```python
# ==============================================================================
#  ขั้นตอนที่ 1: ติดตั้งไลบรารีและ Imports
# ==============================================================================
import os
print("⏳ กำลังติดตั้งไลบรารี ultralytics...")
os.system('pip install -q ultralytics torchvision')
print("✅ ติดตั้งเสร็จสิ้น!")

import torch
import torch.nn as nn
from ultralytics import YOLO
from torchvision.models import mobilenet_v2

# ==============================================================================
#  ขั้นตอนที่ 2: สร้างโมดูล Backbone ใหม่ (สำคัญที่สุด)
# ==============================================================================
# เราจะสร้าง Class ใหม่ที่ห่อหุ้ม MobileNetV2 ไว้
# และสั่งให้มันส่ง Feature Map ที่เราต้องการออกมา 3 ระดับ (P3, P4, P5)

class MobileNetV2Backbone(nn.Module):
    def __init__(self, pretrained=True):
        super().__init__()
        # โหลด MobileNetV2 ที่ train มาแล้วบน ImageNet
        mobilenet = mobilenet_v2(weights='IMAGENET1K_V1' if pretrained else None)

        # เราต้องการ Feature Map จาก Layer ที่ 6, 13, และ 17
        self.features = mobilenet.features
        self.feature_indices = (6, 13, 17)

    def forward(self, x):
        # สร้าง list เพื่อเก็บ outputs ที่เราต้องการ
        outputs = []
        # วนลูปผ่านแต่ละ layer ของ MobileNetV2
        for i, module in enumerate(self.features):
            x = module(x)
            # ถ้าเจอ layer ที่เราต้องการ ให้เก็บ output ไว้
            if i in self.feature_indices:
                outputs.append(x)
        return outputs

# ==============================================================================
#  ขั้นตอนที่ 3: ทำการ "ผ่าตัด" เปลี่ยน Backbone
# ==============================================================================
print("\n--- 🩺 กำลังเริ่มต้นการผ่าตัดเปลี่ยน Backbone ---")

# 1. โหลดโมเดล YOLOv12m เดิม
yolo_model = YOLO('yolo12m.pt')
original_layers = yolo_model.model.model

# 2. สร้าง Backbone ใหม่ของเรา
new_backbone = MobileNetV2Backbone()

# 3. สร้างโมเดลใหม่โดยเอา Backbone ใหม่ไปแทนที่ส่วน Backbone เดิม
#    Backbone เดิมของ YOLOv12m คือ 10 layers แรก (index 0-9)
#    เราจะแทนที่ 10 layers นี้ด้วย new_backbone ของเราเพียงชิ้นเดียว
new_layers = nn.Sequential(
    new_backbone,  # Backbone ใหม่ของเรา
    *list(original_layers.children())[10:] # เอา Neck + Head เดิมมาต่อท้าย
)

# 4. นำโครงสร้างใหม่ใส่กลับเข้าไปในโมเดล YOLO
yolo_model.model.model = new_layers

# 5. "เดินสายไฟใหม่" ให้กับ Head (แก้ไข `IndexError` ที่นี่)
#    หา Head (โมดูล Detect) ซึ่งตอนนี้จะอยู่ที่ Layer สุดท้าย
detect_head = yolo_model.model.model[-1]

#    Backbone ใหม่ของเราส่ง Output ออกมา 3 ค่าจาก Layer เดียว (Layer 0)
#    ดังนั้นเราต้องบอกให้ Head ไปดึงข้อมูลจาก Layer 0 ทั้ง 3 ค่า
#    (เนื่องจาก PyTorch ไม่สามารถดึงหลาย output จาก layer เดียวได้โดยตรง
#     ultralytics จะจัดการให้เองถ้าเราบอกว่าให้ดึงจาก layer 0)
#    การตั้งค่า f=[0, 0, 0] เป็นวิธีบอกให้ระบบรู้ว่าต้องจัดการเรื่องนี้
#    ในทางปฏิบัติ เราจะเชื่อมกับ layer ก่อนหน้า Neck ที่ทำหน้าที่แยก feature
#    เพื่อให้ง่าย เราจะเชื่อมกับ layer สุดท้ายของ Neck ก่อนหน้า Head
#    ในโครงสร้าง YOLOv8/v12 มาตรฐานคือ layer ที่ 15, 18, 21
detect_head.f = [15, 18, 21]

print("✅ การผ่าตัดสำเร็จ!")

# ==============================================================================
#  ขั้นตอนที่ 4: แสดงผลลัพธ์
# ==============================================================================
print("\n\n--- 🎉 ผลลัพธ์โมเดลใหม่หลังการผ่าตัด ---")
yolo_model.info(detailed=True)
print("\n\nคุณได้สร้างโมเดล YOLO ที่มี MobileNetV2 เป็น Backbone ได้สำเร็จแล้ว!")

# (ใส่ต่อจากโค้ดเดิมที่คุณรันสำเร็จแล้ว)

# ==============================================================================
#  ขั้นตอนที่ 5: บันทึกโมเดลใหม่เป็นไฟล์ .pt
# ==============================================================================
# กำหนดชื่อไฟล์ใหม่
new_model_filename = "yolo12m_mobilenetv2.pt"

# ใช้คำสั่ง .save() เพื่อบันทึกโมเดล
yolo_model.save(new_model_filename)

print(f"\n\n✅ โมเดลถูกบันทึกเรียบร้อยแล้วที่ '{new_model_filename}'")
```

โมเดลนี้ index error น่าจะเป็นเพราะ ultralytics ไม่เข้าใจ torchvision

```python
import torch
import torch.nn as nn
from ultralytics import YOLO
from torchvision.models import mobilenet_v2

# ==============================================================================
#  ขั้นตอนที่ 1: สร้าง Backbone ที่มี "หม้อแปลง" (Adapter) ในตัว
# ==============================================================================
class AdaptedMobileNetV2Backbone(nn.Module):
    def __init__(self, pretrained=True):
        super().__init__()
        mobilenet = mobilenet_v2(weights='IMAGENET1K_V1' if pretrained else None)
        self.features = mobilenet.features
        self.feature_indices = (6, 13, 17)
        yolo_channels_needed = [512, 512, 1024]
        mobilenet_channels_out = [32, 96, 320]
        self.adapters = nn.ModuleList([
            nn.Conv2d(in_ch, out_ch, kernel_size=1)
            for in_ch, out_ch in zip(mobilenet_channels_out, yolo_channels_needed)
        ])

    def forward(self, x):
        outputs = []
        mobilenet_outputs = []
        for i, module in enumerate(self.features):
            x = module(x)
            if i in self.feature_indices:
                mobilenet_outputs.append(x)
        for i, adapter in enumerate(self.adapters):
            outputs.append(adapter(mobilenet_outputs[i]))
        return outputs

# ==============================================================================
#  ขั้นตอนที่ 2: ทำการ "ผ่าตัด" และเตรียม "รายการชิ้นส่วน"
# ==============================================================================
print("--- 🩺 กำลังสร้างโมเดล YOLO ที่มี MobileNetV2 จริงๆ ---")

# 1. โหลดโมเดลเดิมเพื่อ "คัดลอก" ส่วน Neck+Head เก็บไว้
yolo_original = YOLO('yolo12m.pt')
original_neck_head = list(yolo_original.model.model.children())[10:]

# 2. สร้างโมเดลสุดท้ายที่เราจะใช้งาน และทำการผ่าตัด
final_model = YOLO('yolo12m.pt')
new_adapted_backbone = AdaptedMobileNetV2Backbone()
final_model.model.backbone = new_adapted_backbone
print("✅ การผ่าตัดสำเร็จ!")

# 3. สร้าง "รายการชิ้นส่วนที่ถูกต้อง" สำหรับการแสดงผล
modules_to_display = [final_model.model.backbone] + original_neck_head


# ==============================================================================
#  ขั้นตอนที่ 3: X-Ray โครงสร้างใหม่ (***วิธีที่ถูกต้องที่สุด***)
# ==============================================================================
print("\n--- 🔬 กำลัง X-Ray โครงสร้างและผลลัพธ์ของโมเดลใหม่ ---")
dummy_input = torch.rand(1, 3, 640, 640)

# 1. ตรวจสอบ "ผลผลิต" ที่ออกจาก Backbone ใหม่ของเราโดยตรง
backbone_output = final_model.model.backbone(dummy_input)
backbone_output_shapes = [o.shape for o in backbone_output]

# 2. ตรวจสอบ "ผลผลิตสุดท้าย" ที่ออกจากโมเดลทั้งหมด
final_output = final_model(dummy_input)

print("\n--- 📝 รายงานผลการ X-Ray (โครงสร้างระดับบล็อก) ---")
# พิมพ์ "รายการชิ้นส่วน" ที่เราใช้ประกอบโมเดล
for i, module in enumerate(modules_to_display):
    module_name = module.__class__.__name__
    if i == 0: # ถ้าเป็น Backbone ใหม่ของเรา
        shape_str = str(backbone_output_shapes).replace('torch.Size(', '').replace(')', '')
        print(f"Block {i:<2}: {module_name:<25} \t Output Shape: {shape_str}")
    else: # สำหรับ Neck และ Head เราจะแสดงแค่ชื่อบล็อกและสายไฟ
        from_str = f"\n  └── รับข้อมูลมาจาก Layer: {module.f}" if hasattr(module, 'f') else ""
        print(f"Block {i:<2}: {module_name:<25} \t {from_str}")


# ==============================================================================
#  ขั้นตอนที่ 4: แสดงผล .info() ซึ่งจะแสดง "ชิ้นส่วนย่อย" ทั้งหมด
# ==============================================================================
print("\n\n--- 🎉 ผลลัพธ์โมเดลใหม่หลังการผ่าตัด (มุมมองระดับชิ้นส่วนย่อย) ---")
final_model.info(detailed=True)
print("\n\n🎉🎉🎉 สำเร็จ! คุณได้สร้างและตรวจสอบ YOLO ด้วย MobileNetV2 จริงๆ ได้แล้ว! 🎉🎉🎉")

# ==============================================================================
#  ขั้นตอนที่ 5: บันทึกโมเดลใหม่เป็นไฟล์ .pt (***ส่วนที่เพิ่มเข้ามา***)
# ==============================================================================
new_model_filename = "yolo12m_adapted_mobilenetv2.pt"
final_model.save(new_model_filename)
print(f"\n\n💾 โมเดลไฮบริดถูกบันทึกเรียบร้อยแล้วที่ '{new_model_filename}'")
print("คุณสามารถนำไฟล์นี้ไปใช้เทรนต่อได้ทันที!")
```
